```python
## Homer Age effect
results = {}
for s in ['younger', 'older']:
    f = os.path.join(homer_dir, s, 'homerResults.html')
    soup = BeautifulSoup(open(f, 'rb'), 'html.parser')

    strs = np.array([x.string for x in soup.find_all('td')])
    mtch = np.array([x for x in soup.find_all('td')])

    br = np.where(strs == '1\n')[0][0]
    cols = strs[:br]
    data = strs[br:]
    data = data.reshape([int(len(data)/br), br])

    selected = np.fromiter(range((2*br)-2,br*data.shape[0]+br,br),dtype=int)
    TFs = [x.get_text().split('/')[0] for x in mtch[selected]]

    results[s] = pd.DataFrame(data, columns=cols)
    TFs = np.array([x.split('.')[0] for x in TFs])
    results[s]['Transciption Factor'] = [x.split('(')[0] if len(x) > 20  else x for x in TFs]
    results[s]['log P-pvalue'] = - results[s]['log P-pvalue'].astype(float)
    results[s]['P-value'] = results[s]['P-value'].astype(float)
    results[s] = results[s].sort_values(by='log P-pvalue', ascending=True)

    
mx = np.max(np.hstack([results['younger']['log P-pvalue'],results['older']['log P-pvalue']]))

fig, ax = plt.subplots(1,2,figsize=(6,4))
results['younger'].plot.barh(y = 'log P-pvalue', x = 'Transciption Factor',
             legend = None, ax = ax[0], color = 'lightblue', ylabel = None,
             fontsize=8);
ax[0].tick_params(labelright=True, labelleft=False, labelsize=8);
ax[0].yaxis.tick_right()
for pos in ['left', 'top']:
    ax[0].spines[pos].set_visible(False)
    
results['older'].plot.barh(y = 'log P-pvalue', x = 'Transciption Factor',
         legend = None, ax = ax[1], color = 'orange', ylabel = None,
         fontsize=8);
for pos in ['right', 'top']:
    ax[1].spines[pos].set_visible(False);
    
ax[0].set_xlabel('-log p-value')
ax[1].set_xlabel('-log p-value')
ax[0].set_ylabel(None)
ax[1].set_ylabel(None)

# ax[0].set_xlim(mx, 0);
# ax[1].set_xlim(0, mx);
ax[0].set_xlim(np.max(results['younger']['log P-pvalue']), 0);
ax[1].set_xlim(0, np.max(results['older']['log P-pvalue']));

plt.subplots_adjust(wspace=.8, hspace=0)
plt.savefig(os.path.join(export_folder, 'fig2_Age_TFs.svg'), bbox_inches='tight')
```

```python
## Topic modelling
def plot_modules(ds, emb:str='TSNE', plot_name:str='modules.png', titles:bool=True):
    modules = ds.ra.module
    names, labels = np.unique(modules, return_inverse=True)
    fig, ax = plt.subplots(1,2, figsize=(2,1.33))
    
    plt.subplot(1,2,1)
    scatterc(ds.ra[emb], c=labels, legend=None);
    if titles:
        plt.title('Regulatory Modules', fontsize=6)
    plt.axis('off')
    plt.gca().set_aspect('equal')
    
    d = abs(ds.ra['Distance to TSS'].astype(int))
    colors = ['#b30000','#fc8d59','#fef0d9']
    tag_names = ['TSS', 'Proximal', 'Distal']
    tag = np.zeros(len(d)).astype(int)
    tag[np.where((200<d) & (d<2000))[0]] = 1
    tag[np.where(d>2000)[0]] = 2
    
    plt.subplot(1,2,2)
    plt.scatter(ds.ra[emb][:,0], ds.ra[emb][:,1], c=[colors[i] for i in tag], lw=0, marker='.', s=.5)
    if titles:
        plt.title('TSS distances', fontsize=6)
    plt.axis('off')
    plt.gca().set_aspect('equal')

    plt.savefig(plot_name, dpi=1200, bbox_inches='tight')
    
with loompy.connect(f_peaks_agg) as ds:
    plot_modules(ds, emb='TSNE', plot_name= os.path.join(export_folder, 'fig2_cCRE_modules.png'))
```

```python
def motif_plot(ds: loompy.LoomConnection, out_file: str, label:str = 'TF', label2: str = None, N:int = None, threshold:int = None, TFs: list = None, N_cluster: int = None, must_include: list = None, text: bool = True) -> None:
    '''
    Generates a multi-panel plot to inspect Motif enrichment scores.
    
    Args:
        ds                    Connection to the .agg.loom motif file to use
        dsr                   Connection to the .agg.loom RNA file to use
        out_file              Name and location of the output file
        N                     The number of motifs to include per cluster (top N)
        
    Remarks:
    
    '''
    layer = '-log_pval'
    if TFs:
        valids = np.isin(ds.ra.Cluster, TFs)
    elif N:
        valids = []
        for i in range(ds.shape[1]):
            q = np.quantile(ds[layer][:,i], (ds.shape[0]-N)/ds.shape[0])
            valids.append(np.where(ds[layer][:,i]>q)[0])
        valids = np.unique([i for s in valids for i in s])
    else:
        valids = np.repeat(True,ds.shape[0])
    ds.ra.valids = np.zeros(ds.shape[0])
    ds.ra.valids[valids] = True
        
    mask = np.zeros(ds.shape[0], dtype=bool)
    mask[valids] = 1
    
    TF_order = np.zeros(ds.shape[0], dtype='int')
    TF_order[mask] = np.argmax(ds.layer[layer][np.where(mask)[0], :], axis=1)
    TF_order[~mask] = np.argmax(ds.layer[layer][np.where(~mask)[0], :], axis=1) + ds.shape[1]
    TF_order = np.argsort(TF_order)[::-1]
    ds.permute(TF_order, axis=0)
    
    shape_factor = len(valids)/ds.shape[1]

    layer = '-log_pval'
    x = np.where(ds.ra.valids)[0]
    pval = ds[layer][x,:]

    df = pd.DataFrame([])
    if label2:
        TF = [f'{ds.ra[label][i]}:{ds.ra[label2][i]}' for i in x]
    else:
        TF = [f'{ds.ra[label][i]}' for i in x]
    
    Topic_d = {}
    for i in range(ds.shape[1]): 
        pvals = ds[layer][:,i][x]
        data = pd.DataFrame({'Topic': i, 'label': ds.ca.Topic[i], 'TF': TF, 'p_vals': pvals}, columns=['Topic', 'label','TF', 'p_vals'])
        df = df.append(data)
        
    factor = np.max(df['p_vals']) / 25
    df['Bubble_size'] = df['p_vals'] / factor
    order = sorted(np.unique(df['Topic']))
    df = df.reset_index(drop=True) 

    if N_cluster:
        N_cluster += 1
        valid_Topics = []
        for i in np.unique(df['TF']):
            vals = np.array(df.loc[df['TF'] == i,'p_vals'])
            q = vals[np.argsort(vals)][-N_cluster]
            valid_Topics.append(df['Topic'][(df['TF']== i)&(df['p_vals']>q)].values)
            
        if must_include:
            valid_Topics.append(np.where(np.isin(ds.ca.Topic, must_include))[0])
        valid_Topics = np.unique([x for s in valid_Topics for x in s])
        df = df[df['Topic'].isin(valid_Topics)]
        df = df.reset_index(drop=True) 

    if threshold:
        df = df[df['p_vals'] > threshold] 
        df = df.reset_index(drop=True)

    d_order = {k:v for v,k in enumerate(TF)}
    df['order'] = np.array([d_order[x] for x in df['TF']])
    
    d_order = {k:v for v,k in enumerate(sorted(np.unique(df['Topic'])))}
    df['Topic2'] = np.array([d_order[x] for x in df['Topic']])
    
    d_rev = {}
    for i in df.index:
        d_rev[df['Topic2'][i]] = df['label'][i]
    
    df = df.sort_values(ascending=True, by='order')
    nrow, ncol = len(np.unique(df['TF'])), len(np.unique(df['Topic']))
    h, w, = 1.2,4
    fig, ax = plt.subplots(figsize=(w,h))

    scatter = ax.scatter('Topic2', 'TF', s='Bubble_size', c='black', data=df, edgecolors='black', linewidth=.5, label=None)

    xlabels = [d_rev[i] for i in range(np.max(df['Topic2'])+1)]
    if text:
        for p in [250,500,750,1000]:
            plt.scatter([], [], c='k', alpha=0.3, s=p/factor, label=str(p), lw=0)
        plt.legend(scatterpoints=1, frameon=False, labelspacing=1, bbox_to_anchor=(1.01,1),
                   title='-log(p)', title_fontsize=6, fontsize=4)
        plt.xticks(np.arange(len(xlabels)), xlabels, rotation=70,  fontsize=5) 
        plt.yticks(range(len(TF)), TF, fontsize=5) 
    else:
        plt.xticks([])
        plt.yticks([])
    plt.xlim(-1, ncol)
    plt.ylim(-1, nrow)
    plt.savefig(out_file, bbox_inches='tight', dpi=600)
```
