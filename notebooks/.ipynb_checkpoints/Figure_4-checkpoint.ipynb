```python
import os, glob
import matplotlib.pyplot as plt
import loompy
import numpy as np
import pandas as pd

from cytograph.visualization.scatter import *

from pybedtools import BedTool
import torch
from deeplift.dinuc_shuffle import dinuc_shuffle
from deeplift.visualization import viz_sequence
from twobitreader import TwoBitFile
from captum.attr import DeepLiftShap

from fetalxbrainxmultiomics.deepNeuron.model import ConvolutionalClassificationModel
from fetalxbrainxmultiomics.deepNeuron.utils import *

def moving_average(data, window:int = 10):
    kernel = np.ones(window) / window
    return np.convolve(data, kernel, mode='same')
```

```python
from pygam import PoissonGAM, LinearGAM, s

genes = ['PTF1A', 'ASCL1', 'NEUROG2', 'TFAP2B', 'NHLH1', 'NHLH2', 
       'LHX5', 'PAX2', 'EBF3', 'FOXP2', 'ESRRB', 'RORA', 'LHX1', 'PCP4']

## First we fit a GAM with poisson to the gene expression data
with loompy.connect(f_RNA) as ds:
    valid = np.array([np.where(ds.ra.Gene == g)[0][0] for g in genes])
    
    X = np.arange(ds.shape[1])
    X2 = np.arange(0,ds.shape[1], 100)
    data = []
    for i in valid:
        Z = ds[i,:].flatten()
        if np.sum(Z>0) > 100:
            GAM_RNA = PoissonGAM(s(0, n_splines=10)).fit(X, Z)
            data.append(GAM_RNA.predict(X2))
    data = np.array(data)
    
    data = (data - np.min(data,1)[:,np.newaxis]) / (np.max(data,1) - np.min(data,1))[:,np.newaxis]
    
    plt.figure(figsize=(4,.5*data.shape[0]))
    plt.imshow(data, aspect='auto', interpolation='nearest', cmap='Reds')
    plt.yticks(np.arange(len(genes)), labels=genes)
    plt.xticks([])
    
plt.savefig(os.path.join(export_folder, 'fig4_marker_heatmap.svg'), bbox_inches='tight')
```

```python
## For the predicted expression in the simulated data we used a GAM with a normal distribution (LinearGAM) 
## since the predicted values are continuous and not discrete

data = []
t = np.stack(exp.columns.str.split('_').values)[:,1].astype(int)
for g in genes:
    GAM = LinearGAM(s(0, n_splines = 10)).fit(t, exp.loc[g])
    data.append(GAM.predict(np.arange(np.max(t))))
data = np.array(data)

plt.figure(figsize=(4,.5*data.shape[0]))
plt.imshow(data, aspect='auto', interpolation='nearest', cmap='Oranges')
plt.yticks(np.arange(len(genes)), labels=genes)
plt.xticks([])

plt.savefig(os.path.join(export_folder, 'fig4_prediction_heatmap.svg'), bbox_inches='tight')
```

```python
## Genes - enhancers - ESRRB - Targets
from cytograph.visualization.colors import colorize
from pygam import LinearGAM, LogisticGAM, PoissonGAM, s

w, h = 8, 10

genes_upstream = ['TFAP2B', 'LHX5']
gene_target = 'ESRRB'
step = 10
step2 = 100 ## Apply larger smoothing window to cCREs due to the sparse, near binary nature of scATAC-seq

# colors_upstream = ['#2bcd48', '#80baed', '#fe5005']
colors_upstream = ['#24408e', '#fe5005']

f_gene_links = '../files/Figure_4/Gene_links.bedpe'
links = BedTool(f_gene_links)

with loompy.connect(f_peaks) as ds:
    with loompy.connect(f_RNA) as dsr:
        X = np.arange(ds.shape[1])
        Z_upstreams = []
        Z_pred_upstreams = []
        
        ## Upstream Genes GAM
        for g in genes_upstream:
            Z = dsr[dsr.ra.Gene==g,:].flatten()
            GAM_RNA = LinearGAM(s(0, n_splines=10)).fit(np.arange(len(Z)), Z)
            Z_pred = GAM_RNA.predict(X)
            mx = np.max(Z_pred)
            Z_pred_upstreams.append(Z_pred / mx)
            Z_upstreams.append(moving_average(Z, step) / mx)
        
        ## Enhancers GAM
        valid = BedTool([x[3:6] for x in links if (x[6] == gene_target)])
        valid = BedTool([[ds.ra.Chr[i],ds.ra.Start[i], ds.ra.End[i],ds.ra.ID[i]] for i in range(ds.shape[0])]).intersect(valid)
        valid = [x[3] for x in valid]
        valid = np.isin(ds.ra.ID, valid) & ['exon' not in x for x in ds.ra.Annotation] & ['TSS' not in x for x in ds.ra.Annotation]
        cCRE = np.where(valid)[0]
        enh = ds.ra.ID[cCRE]
        factor = ds.ca.passed_filters / 5000
        
        Y_enh = []
        Y_pred_enh = []
        for i, x in enumerate(cCRE):
            v = ds[x,:].flatten()
            Y = v / factor
            GAM_ATAC = LinearGAM(s(0, n_splines=10)).fit(np.arange(len(Y)), Y)
            Y_pred = GAM_ATAC.predict(X)
            mx = np.max(Y_pred)
            Y_pred_enh.append(Y_pred / mx)
            Y_enh.append(moving_average(Y, step2))
            
        Y_pred_enh = np.array(Y_pred_enh)
            
        ## Fit target GAM
        Z = dsr[dsr.ra.Gene==gene_target,:].flatten()
        GAM_RNA = LinearGAM(s(0, n_splines=10)).fit(np.arange(len(Z)), Z)
        Z_pred = GAM_RNA.predict(X)
        mx = np.max(Z_pred)
        Z_gene_pred = Z_pred / mx
        Z_gene_smooth = moving_average(Z, step) / mx
        
        fig, ax = plt.subplots(4,1, figsize=(w,h))
        
        plt.subplot(4,1,1)
        colors = colorize(genes_upstream)        
        for i, g in enumerate(genes_upstream):
            plt.plot(Z_pred_upstreams[i], label = g, c=colors_upstream[i])
            plt.scatter(np.arange(len(Z_upstreams[i])), Z_upstreams[i], s=.5, lw=0, alpha=.1, c=colors_upstream[i]);
        plt.ylim((0,1.25))
        plt.xticks([])
        plt.yticks([])
        plt.title('Expression of genes regulating ESRRB')
        
        plt.subplot(4,1,2)
        plt.xticks([])
        plt.yticks([])
        colors = []
        for i in range(Y_pred_enh.shape[0]):
            if Y_pred_enh[i,15000] > .5:
                colors.append('#80baed')
            else:
                colors.append('#ffa782')
        for i, x in enumerate(enh):
            plt.plot(Y_pred_enh[i], lw=1, c=colors[i], label=x)
        plt.title(f'cCRE upstream of ESRRB')
            
        plt.subplot(4,1,3)
        plt.plot(Z_gene_pred, lw=3, c='#9ccb00')
        plt.xticks([])
        plt.yticks([])
        plt.scatter(np.arange(len(Z_gene_smooth)), Z_gene_smooth, s=.5, lw=0, alpha=.1, c='#9ccb00');
        plt.ylim((0,1.25))
        plt.title('ESRRB Expression')
        
with loompy.connect(f_motifs) as ds:
    Z = ds['Z'][ds.ra.TF==gene_target,:].flatten()
    GAM_motif = LinearGAM(s(0, n_splines=10)).fit(np.arange(len(Z)), Z)
    Z_pred = GAM_motif.predict(X)
    Z_motif_pred = Z_pred
    Z_motif_smooth = moving_average(Z, step)
    mx = np.quantile(Z_motif_smooth, .95)
    
    plt.subplot(4,1,4)
    plt.plot(Z_motif_pred, lw=3, c='#e080c3')
    plt.xticks([])
    plt.yticks([])
    plt.scatter(np.arange(len(Z_motif_smooth)), Z_motif_smooth, s=.5, lw=0, alpha=.1, c='#e080c3');
    plt.ylim((-mx, mx))
    plt.title('ESRRB motif enrichment')
    
plt.savefig(os.path.join(export_folder, 'fig4_dynamics.png'), dpi=600, bbox_inches='tight')
```

```python
## Plot deepNeuron nucleotide enrichments

## Load the model
model = ConvolutionalClassificationModel(n_celltypes=5, length=401)
model.load_state_dict(state_dict = torch.load(f_model))
model.eval()

f_bit = '/datb/sl/camiel/scATAC/ref/hg38.2bit' ## The .2bit reference can be downloaded from here: https://hgdownload-test.gi.ucsc.edu/goldenPath/hg38/bigZips/hg38.2bit

## Enhancers to annotate
enhancers = ['chr14:76073664-76074065', 'chr14:76076503-76076904']
sequences = one_hot_encoder(enhancers, 401, f_bit)

hypo_contributions = []
contrib_scores= []

explainer = DeepLiftShap(model)

for i in trange(len(sequences)):
    onehot_data = sequences[i]
    mult = explainer.attribute(torch.tensor(onehot_data[None, :, :]), shuffle_several_times(onehot_data), target=4).detach().numpy()[0]
    bg_data = np.mean(shuffle_several_times(onehot_data).numpy(),axis=0)

    hypo = combine_mult_and_diffref(mult, orig_inp=torch.tensor(onehot_data[None, :, :]), bg_data=bg_data)
    hypo_contributions.append(hypo)
    
    dinuc_shuff_explanation = np.sum(mult, axis=0) * onehot_data
    contrib_scores.append(dinuc_shuff_explanation)

nuc_colors = ['Green', 'Blue', 'Orange', 'Red']
for i in range(2):
    fig, ax = plt.subplots(1,1, figsize=(20,2))
    plt.title(enhancers[i])
    viz_sequence.plot_weights_given_ax(ax, contrib_scores[i], .1, .1, highlight = [], subticks_frequency=20)
    
    hypo = hypo_contributions[i].T
    X = np.arange(hypo.shape[0])
    plt.figure(figsize=(20,2))
    for x in range(4):
        plt.scatter(X, hypo[:,x], c=nuc_colors[x], lw=0, s=2)
    plt.xticks(np.arange(0,hypo.shape[0],20))
    mx = np.max(abs(hypo))
    plt.ylim((-mx,mx))
    plt.xlim((0,401))
```
