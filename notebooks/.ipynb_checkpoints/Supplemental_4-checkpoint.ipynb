```python
## Conservation
extend = 300

g_path = '/home/camiel/chromograph/chromograph/references/male.GRCh38.chrom.sizes'
f_phast = '/datb/sl/camiel/scATAC/ref/hg38.phastCons100way.bw'

bg_peaks = BedTool('/proj/DEV_ATAC/submission/enhancer_regions/random_seqs_100k_1001bp.sorted.bed')
n_bg = bg_peaks.count()
peaks = BedTool('/proj/DEV_ATAC/submission/enhancer_regions/consensus_peaks.bed').slop(b=extend, g=g_path)
total_peaks = peaks.count()

bg_vals = np.zeros([n_bg, 1001])
with pyBigWig.open(f_phast) as bw:
    for i, interval in enumerate(bg_peaks):
        bg_vals[i,:] = np.nan_to_num(bw.values(interval[0], int(interval[1]), int(interval[2]), numpy=True),0)
bg = np.mean(bg_vals, 0)

scores = np.zeros([total_peaks, 401 + (2*extend)])
with loompy.connect(f_peaks_agg) as ds:
    with pyBigWig.open(f_phast) as bw:
        for i in range(ds.shape[0]):
            scores[i,:] = np.nan_to_num(bw.values(ds.ra.Chr[i], ds.ra.Start[i].astype(int)-extend, ds.ra.End[i].astype(int)+extend+1, numpy=True),0)
        
    prom = abs(ds.ra['Distance to TSS'].astype(int)) < 2000
    proximal = np.mean(scores[prom,:], 0)
    distal = np.mean(scores[~prom,:], 0)
```

```python
# ## Encode
from tqdm import tqdm
# import cytograph.visualization.colors as colors

peaks1 = BedTool('/proj/DEV_ATAC/submission/enhancer_regions/consensus_peaks.bed')
peaks2 = BedTool('/datb/sl/camiel/tumor/Healthy_tissues_scATAC/Ren_brain/hba.whole.union.peaks.bed')
ALL = BedTool('/proj/DEV_ATAC/submission/enhancer_regions/repeat_masker_GRCh38_filtered.bed')

totals = [x.count() for x in [peaks1, peaks2]]
values = []
mn = 50

for i, peaks in enumerate([peaks1, peaks2]):

    total = totals[i]
    overlaps = peaks.intersect(ALL, wo=True)
    if np.array(overlaps[0]).shape[0] == 11:
        df = pd.read_table(overlaps.fn, names=['chr', 'start', 'end' , 'ID', 'score', 'strand', 'chr2', 'start2', 'end2', 'Element', 'bp'])
    else: 
        df = pd.read_table(overlaps.fn, names=['chr', 'start', 'end' , 'ID', 'chr2', 'start2', 'end2', 'Element', 'bp'])

    L, S, LT, = 0, 0, 0
    for ID in tqdm(np.unique(df['ID'])):
        X = df[df['ID'] == ID]
        if np.max(X['bp']) > mn:
            s = X.iloc[np.argmax(X['bp'])]['Element']
            if s == 'LINE':
                L += 1
            elif s == 'SINE':
                S += 1
            elif s == 'LTR':
                LT += 1

    res = np.zeros(3)
    res[0] = 100 * L / total
    res[1] = 100 * S / total
    res[2] = 100 * LT / total
    values.append(res)

    print(res[0], ' : ', L)
    print(res[1], ' : ', S)
    print(res[2], ' : ', LT)

values = np.array(values)

w = .75
bottom = np.zeros(2)
labels = ['LINE', 'SINE', 'LTR']
colours = ['#e4bf80', '#f19670', '#5698c4']

fig, ax = plt.subplots(figsize=(2,5))
for i in range(values.shape[1]):
    ax.bar(np.array([1,2]), values[:,i], bottom = bottom, color = colours[i], lw=.75, edgecolor='black', width=w)
    bottom = bottom + values[:,i]
           
ax.bar(np.array([1,2]), 100 - bottom, bottom = bottom, color = 'lightgrey', lw=.75, edgecolor='black', width=w)
ax.bar(np.array([1,2]), .5, bottom = 100, color = 'white', lw=0, width=1)
# ax.set_yticklabels(['0%', '20%', '40%', '60%', '80%', '100%'])
ax.set_yticklabels([])
# ax.set_ylabel('Percentage of elements in \n published datasets identified')
ax.set_xticks([])

# plt.savefig('/datb/sl/camiel/scATAC/DEV_manuscript/sup_4/Transposable_elements.png', bbox_inches='tight', dpi=600)
```

```python
from cytograph.visualization.colors import Colorizer

with loompy.connect('/proj/DEV_ATAC/pycistopic_test/DEV_downsample/Pool_downsampled.loom') as ds:
    region_topics = ds.ca.topics.T
    clusters = ds.ca.Clusters
topics = region_topics / np.max(region_topics, 0)[np.newaxis,:]
# topics = region_topics / np.sum(region_topics, 0)[np.newaxis,:]
# topics = region_topics

colors = Colorizer(scheme='colors75').fit_transform(clusters)

sum_topic = []
for  i in range(135):
    sum_topic.append(np.mean(topics[:,clusters==i], 1))
sum_topic = np.array(sum_topic)

order = np.argmax(sum_topic, axis=0)
order = np.argsort(order)
values = topics[order,:]

fig, ax = plt.subplots(2,1, figsize=(13,6), gridspec_kw={'height_ratios': [.5,12]})

ax[0].bar(np.arange(values.shape[1]), 1, color = colors, width=1)
ax[0].set_xlim([0,values.shape[1]])
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[1].imshow(values,interpolation='none', aspect='auto');
ax[1].set_xticks([])
ax[1].set_yticks([])

for i in range(135):
    x = np.where(clusters == i)[0][-1]
    ax[1].axvline(x, c='white', lw=.25)

plt.subplots_adjust(wspace=0, hspace=0.01)

plt.savefig('/datb/sl/camiel/scATAC/DEV_manuscript/sup_4/topics_heatmap.png', dpi=1200, bbox_inches = 'tight')
```

```python
## From Milopy

## Homer Age effect

from bs4 import BeautifulSoup

fdir = '/proj/DEV_ATAC/submission/TSS_topics/homer/'
labels = ['Topic_9', 'Topic_18']
results = {}
for s in labels:
    f = os.path.join(fdir, s, 'homerResults.html')
    soup = BeautifulSoup(open(f, 'rb'), 'html.parser')

    strs = np.array([x.string for x in soup.find_all('td')])
    mtch = np.array([x for x in soup.find_all('td')])

    br = np.where(strs == '1\n')[0][0]
    cols = strs[:br]
    data = strs[br:]
    data = data.reshape([int(len(data)/br), br])

    selected = np.fromiter(range((2*br)-2,br*data.shape[0]+br,br),dtype=int)
    TFs = [x.get_text().split('/')[0] for x in mtch[selected]]
    
    print(TFs)

#     for i in range(len(TFs)):
#         if '_' in TFs[i]:
#             TFs[i] = '_'.join(TFs[i].split('_')[1:])

    results[s] = pd.DataFrame(data, columns=cols)
    TFs = np.array([x.split('.')[0] for x in TFs])
    results[s]['Transciption Factor'] = [x.split('(')[0] if len(x) > 20  else x for x in TFs]
    results[s]['log P-pvalue'] = - results[s]['log P-pvalue'].astype(float)
    results[s]['P-value'] = results[s]['P-value'].astype(float)
    results[s] = results[s].sort_values(by='log P-pvalue', ascending=True)

mx = np.max(np.hstack([results[labels[0]]['log P-pvalue'],results[labels[1]]['log P-pvalue']]))
fig, ax = plt.subplots(1,2,figsize=(8,4))
results[labels[0]].plot.barh(y = 'log P-pvalue', x = 'Transciption Factor',
             legend = None, ax = ax[0], color = 'blue', ylabel = None,
             fontsize=6);
ax[0].set_xlim(0,mx);
# ax[0].set_xlim(np.max(results['younger']['log P-pvalue']), 0);
ax[0].tick_params(labelright=False, labelleft=True, labelsize=6);
# ax[0].yaxis.tick_right()
for pos in ['right', 'top']:
    ax[0].spines[pos].set_visible(False)
    
results[labels[1]].plot.barh(y = 'log P-pvalue', x = 'Transciption Factor',
         legend = None, ax = ax[1], color = 'red', ylabel = None,
         fontsize=6);
ax[1].set_xlim(0,mx);
ax[1].tick_params(labelright=False, labelleft=True, labelsize=6);
for pos in ['right', 'top']:
    ax[1].spines[pos].set_visible(False);
    
ax[0].set_xlabel('-log p-value')
ax[1].set_xlabel('-log p-value')
ax[0].set_ylabel(None)
ax[1].set_ylabel(None)

plt.subplots_adjust(wspace=.6, hspace=0)
plt.savefig('/datb/sl/camiel/scATAC/DEV_manuscript/sup_4/CTCF_topics.svg', bbox_inches='tight')
```
