```python
import os, sys, glob
import pandas as pd
import matplotlib.pyplot as plt
import loompy
import numpy as np
from sklearn.neighbors import NearestNeighbors

from chromograph.pipeline import config
from chromograph.preprocessing.utils import *
from cytograph.plotting.colors import colors75
from cytograph.visualization.scatter import *
from cytograph.visualization.colors import Colorizer
from cytograph.plotting.dendrogram import dendrogram
from cytograph.pipeline.punchcards import (Punchcard, PunchcardDeck, PunchcardSubset, PunchcardView)

import logging
logger = logging.getLogger()
logging.basicConfig(
    format='%(asctime)s %(levelname)-8s %(message)s',
    level=logging.INFO,
    datefmt='%H:%M:%S')

config = config.load_config()
config.paths.build = '/proj/DEV_ATAC/Build_20220906_classes/'

f_peaks = '/proj/DEV_ATAC/Build_20220906_classes/Pool/Pool_peaks.loom'
f_RNA = '/proj/DEV_ATAC/Build_20220906_classes/Pool/Pool_RNA.loom'
```

```python
out_file = '/datb/sl/camiel/scATAC/DEV_manuscript/sup_3/subsets_UMAP.png'

emb = 'UMAP'
ncol = 4
nrow = 2

deck = PunchcardDeck(config.paths.build)
leaves = deck.get_leaves()

fig, ax = plt.subplots(nrow, ncol, figsize=(3*ncol,3*nrow))

names = {'All_Immune': 'Immune',
        'All_FBL': 'Fibroblasts',
        'All_Vascular': 'Vascular',
        'All_RGL': 'Radial Glia and Glioblasts',
        'All_OPC': 'Oligodendrocyte Precursors',
        'All_Neuron_Peptidergic': 'Minor Neurotransmitters',
        'All_Neuron_GABA': 'GABAergic Neurons',
        'All_Neuron_GLUT': 'Glutamatergic Neurons'}

for i, l in enumerate(leaves):
    s = l.longname()
    with loompy.connect(os.path.join(config.paths.build, s, f'{s}_peaks.loom')) as ds:
        
        XY = ds.ca[emb]
        min_pts = 50
        eps_pct = 60
        nn = NearestNeighbors(n_neighbors=min_pts, algorithm="ball_tree", n_jobs=4)
        nn.fit(XY)
        knn = nn.kneighbors_graph(mode='distance')
        k_radius = knn.max(axis=1).toarray()
        epsilon = (500 / (XY.max() - XY.min())) * np.percentile(k_radius, eps_pct)
        
        plt.subplot(nrow,ncol,i+1)
#         plt.title(names[s], fontsize=10)
        scatterc(XY, c=ds.ca.regions, colors="regions", s=epsilon, legend=None)
        plt.axis('off')
        
plt.savefig(out_file, bbox_layout='tight', dpi=600)
```

```python
## Accessibility comparisons
from pybedtools import BedTool

peakdir = '/proj/DEV_ATAC/submission/class_comparison_accessibility_2/output/'
df = pd.read_csv('/proj/DEV_ATAC/submission/class_comparison_accessibility_2/df.csv', index_col=0)

d = {0: 'Non-neural',
    1: 'Rgl',
    2: 'Nbl',
    3: 'Neuron',
    4: 'Gbl',
    5: 'OPC'}

files = sorted(glob.glob(peakdir + '*.bed'))
d_age = {k: v for k,v in zip(df['Samples'],df['Age'])}

df_peaks = pd.DataFrame.from_dict({'split': [], 'Samples': [], 'Age': [], 'peaks': []})
df_peaks['split'] = df_peaks['split'].astype(int)
for f in files:
    x = f.split('/')[-1].split('_')
    n = BedTool(f).count()
    df_peaks = df_peaks.append(pd.Series({'split': int(x[2]), 'Samples': x[1], 'Age': d_age[x[1]], 'peaks': n}), ignore_index=True)
df_peaks['class'] = [d[x] for x in df_peaks['split']]

import scipy.stats as stats

for i in range(6):
    print('%.0f' % np.mean(df_peaks[df_peaks['split']==i]['peaks']))
    
for i in [2,3,4]:
    print(i, ' : ', stats.ttest_ind(df_peaks[df_peaks['split']==1]['peaks'],
                    df_peaks[df_peaks['split']==i]['peaks']))
    
fig, ax = plt.subplots(figsize=(4,4))
colors = ["lightgrey",
"#27b35d",
"#5d25c4",
"#5384db",
'#447c69',
'#9cba19']
boxprops={'linewidth':.5}
medianprops = {'color':'black', 'linewidth':.5}
bplots = []
for i in range(6):
    bplots.append(ax.boxplot(df_peaks[df_peaks['split']==i]['peaks'],positions=[i], widths=[.7], patch_artist=True, 
                             boxprops = boxprops, medianprops=medianprops, whiskerprops=boxprops, capprops=boxprops))

for i, bplot in enumerate(bplots):
    for patch in bplot['boxes']:
        patch.set_facecolor(colors[i])
    
plt.ylim(0,175000);
plt.yticks(np.arange(0,190000, 50000),np.array(np.arange(0,190000, 50000)/1000).astype(int));
plt.ylabel('Number of accessible regions x $10^3$', fontdict={'fontsize':12})
plt.xticks(np.arange(6),np.array(['Non-neural', 'Radial glia', 'Neuroblast', 'Neuron', 'Glioblast', 'OPC']), rotation=20, fontsize=12, ha='right');
plt.savefig('/datb/sl/camiel/scATAC/DEV_manuscript/sup_3/accessiblity_comparison_classes.svg', bbox_inches='tight')
```

```python
## Homer Age effect

from bs4 import BeautifulSoup

fdir = '/proj/DEV_ATAC/submission/class_comparison_accessibility_2/homer_results'
results = {}
for s in ['younger', 'older']:
    f = os.path.join(fdir, s, 'homerResults.html')
    soup = BeautifulSoup(open(f, 'rb'), 'html.parser')

    strs = np.array([x.string for x in soup.find_all('td')])
    mtch = np.array([x for x in soup.find_all('td')])

    br = np.where(strs == '1\n')[0][0]
    cols = strs[:br]
    data = strs[br:]
    data = data.reshape([int(len(data)/br), br])

    selected = np.fromiter(range((2*br)-2,br*data.shape[0]+br,br),dtype=int)
    TFs = [x.get_text().split('/')[0] for x in mtch[selected]]
    
    print(TFs)

#     for i in range(len(TFs)):
#         if '_' in TFs[i]:
#             TFs[i] = '_'.join(TFs[i].split('_')[1:])

    results[s] = pd.DataFrame(data, columns=cols)
    TFs = np.array([x.split('.')[0] for x in TFs])
    results[s]['Transciption Factor'] = [x.split('(')[0] if len(x) > 20  else x for x in TFs]
    results[s]['log P-pvalue'] = - results[s]['log P-pvalue'].astype(float)
    results[s]['P-value'] = results[s]['P-value'].astype(float)
    results[s] = results[s].sort_values(by='log P-pvalue', ascending=True)

    
mx = np.max(np.hstack([results['younger']['log P-pvalue'],results['older']['log P-pvalue']]))

fig, ax = plt.subplots(1,2,figsize=(6,4))
results['younger'].plot.barh(y = 'log P-pvalue', x = 'Transciption Factor',
             legend = None, ax = ax[0], color = 'lightblue', ylabel = None,
             fontsize=8);
ax[0].tick_params(labelright=True, labelleft=False, labelsize=8);
ax[0].yaxis.tick_right()
for pos in ['left', 'top']:
    ax[0].spines[pos].set_visible(False)
    
results['older'].plot.barh(y = 'log P-pvalue', x = 'Transciption Factor',
         legend = None, ax = ax[1], color = 'orange', ylabel = None,
         fontsize=8);
for pos in ['right', 'top']:
    ax[1].spines[pos].set_visible(False);
    
ax[0].set_xlabel('-log p-value')
ax[1].set_xlabel('-log p-value')
ax[0].set_ylabel(None)
ax[1].set_ylabel(None)

# ax[0].set_xlim(mx, 0);
# ax[1].set_xlim(0, mx);
ax[0].set_xlim(np.max(results['younger']['log P-pvalue']), 0);
ax[1].set_xlim(0, np.max(results['older']['log P-pvalue']));

plt.subplots_adjust(wspace=.8, hspace=0)
plt.savefig('/datb/sl/camiel/scATAC/DEV_manuscript/sup_3/Age_TFs.svg', bbox_inches='tight')
```

## RGL / GBL Comparison

```python
regions = [['Telencephalon'], ['Midbrain'], ['Hindbrain', 'Pons', 'Medulla']]

df = pd.DataFrame([])
with loompy.connect(f_peaks) as ds:
    with loompy.connect(f_RNA) as dsr:
        GBL = (dsr[np.where(dsr.ra.Gene == 'TNC'),:]>0) & (dsr[np.where(dsr.ra.Gene == 'BCAN'),:] > 0)
        RGL = (dsr[np.where(dsr.ra.Gene == 'HES1'),:]>0) & ~GBL
        RGL, GBL = RGL.flatten(), GBL.flatten()
        age = np.round(ds.ca.Age).astype(int).flatten()
        for region in regions:
            for a in sorted(np.unique(age)):
                GBL_count = np.sum(GBL & (age == a) & np.isin(ds.ca.regions, region))
                RGL_count = np.sum(RGL & (age == a) & np.isin(ds.ca.regions, region))
                tot = np.sum((age == a) & np.isin(ds.ca.regions, region))
                data = pd.DataFrame({'Region': region[0], 'Age': a, 'GBL': GBL_count, 'RGL': RGL_count, 'Total': tot}, columns=['Region', 'Age', 'GBL', 'RGL', 'Total'], index=[0])
                df = df.append(data)

        df.reset_index()
```

```python
from cytograph.visualization.colors import *

colors = Colorizer(scheme='classes').dict()

for region in np.unique(df['Region']):

    GBL = df.loc[df['Region'] == region, 'GBL'].values
    RGL = df.loc[df['Region'] == region, 'RGL'].values
    tot = df.loc[df['Region'] == region, 'Total'].values
    
    plt.figure()
    plt.title(region)
    plt.plot(df.loc[df['Region'] == region, 'Age'].values, GBL / tot, c=colors['Glioblast'])
    plt.plot(df.loc[df['Region'] == region, 'Age'].values, RGL / tot, c=colors['Radial glia'])
    plt.ylim((0,.7))
```

```python
from cytograph.visualization.colors import *

colors = Colorizer(scheme='classes').dict()

region_list = ['Telencephalon', 'Midbrain', 'Hindbrain']
G = []
R = []

for region in region_list:

    GBL = np.mean(df.loc[df['Region'] == region, 'GBL'].values)
    RGL = np.mean(df.loc[df['Region'] == region, 'RGL'].values)
    tot = np.sum([RGL, GBL])
    
    G.append(GBL / tot)
    R.append(RGL / tot)
    
R = np.array(R)
G = np.array(G)
plt.figure()

fig, ax = plt.subplots(figsize=(2,2))
b1 = ax.bar(np.arange(len(region_list)), R, bottom = np.zeros(3), lw=.75, edgecolor='black', color=colors['Radial glia'])
b2 = ax.bar(np.arange(len(region_list)), G, bottom = R, lw=.75, edgecolor='black', color=colors['Glioblast'])
    
ax.set_xticks([0,1,2])
ax.set_xticklabels(region_list, rotation = 30, ha='right')
ax.set_yticks([0,.5,1])
ax.set_yticklabels(['0%', '50%', '100%'])
ax.set_ylabel('Percentage of cells')

plt.savefig('/datb/sl/camiel/scATAC/DEV_manuscript/sup_3/RGL_GBL.png', bbox_inches='tight', dpi=600)
```
